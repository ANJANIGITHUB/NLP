```python
import pandas as pd
import numpy as np
import tensorflow as tf
import tensorflow_hub as hub
import tensorflow_text as text
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_excel("AccidentCategoryTrainingData.xlsx")
data.head(10)
```

```python
data.groupby("result").describe()
```

```python
le=LabelEncoder()
data["encoded_result"]=le.fit_transform(data["result"])
```

```python
data.tail(40)
```

```python
# Train Test Split
```

```python
X_train,X_test,y_train,y_test=train_test_split(data["content"],data["encoded_result"],test_size=0.20,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#Now lets import BERT model and get embeding vectors
```

```python
bert_preprocess=hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3",name="bert_preprocess")
bert_encoder   =hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4",name="bert_encoder")
```

```python
#Test
preprocess=bert_preprocess(["Hello World","Nice see you"])
encoder=bert_encoder(preprocess)
```

```python
#Creation Functions to create embeddings
def get_text_embeddings(text):
  preprocess=bert_preprocess(text)
  encoder   =bert_encoder(preprocess)

  return encoder["pooled_output"]
```

```python
get_text_embeddings(["Hello World","Nice see you"])
```

```python
# Model building(we can build functional model here.Other type of model is called Sequential Model)
```

```python
#Input data preprocessing
text_input=tf.keras.layers.Input(shape=(),dtype=tf.string,name='text')
preprocess_output=bert_preprocess(text_input)
encoder_output   =bert_encoder(preprocess_output)
```

```python
#Neural Network Layers Creation
num_classes=4
InputandDropout_Layer=tf.keras.layers.Dropout(0.1,name="Dropout")(encoder_output["pooled_output"])
output_layer=tf.keras.layers.Dense(num_classes, activation='softmax', name="output")(InputandDropout_Layer)

```

```python
#Final Model

model=tf.keras.Model(inputs=[text_input],outputs=[output_layer])
```

```python
model.summary()
```

```python
len(X_train)
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=50)
```

```python
model.evaluate(X_test, y_test)
```
